```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
col_to_drop=["street", "city", "statezip", "country","date","yr_renovated","condition","sqft_lot"]

X = house_price.drop(col_to_drop, axis=1)
y = house_price['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print('Coefficients: \n', model.coef_)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Test Values:", y_test)
print("Predicted Values:",y_pred)
```

```python
coeffecients = pd.DataFrame(model.coef_,X.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
sns.regplot( x=y_test, y=y_pred)
plt.title('Regression Plot ')
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

The above regression curve is almost perfect the datapoints lie on or close to the regression line this shows the accuracy of the model

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('Mean Squared Error:', mse)
print('R-squared:', r2)
```

```python
from sklearn import metrics
import numpy as np

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

A high r2 score here 1 represents strong model i.e variables sufficient for price prediction  

* low values of mse , mae , rmse (single digit values )shows strong training of the data reduced chance of wrong prediction
